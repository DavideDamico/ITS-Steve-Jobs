```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Crea i dati per il plot
x = np.linspace(-3, 5, 150).reshape(1, -1)
y = np.linspace(-3, 5, 120).reshape(-1, 1)
z = np.cos(x) + np.sin(y)
x, y = x.flatten(), y.flatten()

# Carica l'immagine di sfondo
background_img = plt.imread('illusione-ottica-hermann.jpg')  # <-- cambia percorso se necessario

fig2, ax2 = plt.subplots()

# Mostra immagine di sfondo
ax2.imshow(background_img, extent=[-3, 5, -3, 5], alpha=1)

# Plot contour sopra all'immagine
n_levels = 6
ax2.contour(x, y, z, n_levels, colors='green', linestyles='-')
cs = ax2.contourf(x, y, z, n_levels, colors='none',
                  hatches=['.', '/', '\\', None, '\\\\', '*'],
                  extend='lower')

# Legenda centrata
artists, labels = cs.legend_elements(str_format='{:2.1f}'.format)
legend = ax2.legend(artists, labels,
                    handleheight=2,
                    framealpha=1,
                    loc='center',
                    bbox_to_anchor=(0.5, 0.5))

# Cambia il colore del testo della legenda in giallo
for text in legend.get_texts():
    text.set_color("cyan")

# Cambia il colore di sfondo della legenda in viola
legend.get_frame().set_facecolor("yellow")

# Stile degli assi
plt.xticks(fontsize=12, color="yellow")
plt.title('Abominio')
plt.yticks(fontsize=12, color="cyan")
plt.xlabel('Asse X', fontsize=14, color="lightgray")
plt.ylabel('Asse Y', fontsize=14, color="lightgray")
plt.show()

```
