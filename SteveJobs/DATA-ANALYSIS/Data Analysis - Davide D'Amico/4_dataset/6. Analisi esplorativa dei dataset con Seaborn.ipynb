```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mlp
import matplotlib.pyplot as plt
%matplotlib inline
# plt.show()
```

```python
churnm = pd.read_csv('churn_miss')
```

```python
print(churnm.head(5))
```

```python
sns.countplot(x = 'gender', data = churnm)
```

```python
sns.barplot(x = 'gender', y = 'lasttrans', data = churnm, palette = "rainbow")
```

```python
sns.boxplot(churnm['age'])
```

```python
sns.boxplot(data = churnm)
```

```python
sns.boxplot(churnm['age'], orient = "h")
```

```python
sns.kdeplot(churnm.age)
```

```python
sns.violinplot(churnm['age'], palette='Set2' , orient= 'h' )
```

```python
sns.stripplot(x = "label", y = "age", data = churnm, jitter = True)
```

```python
sns.stripplot(x = "label", y = "age", data = churnm, hue = "gender", jitter = True)
```

```python
sns.stripplot(x = "label", y = "age", data = churnm, hue = "gender", jitter = True, palette = "Set2")
```

```python
sns.jointplot(x = 'age',y = 'lasttrans', data = churnm, kind = 'scatter')
```

```python
sns.jointplot(x = 'age',y = 'lasttrans', data = churnm, kind = 'hex')
```

```python
sns.lmplot(x = 'lasttrans', y = 'age', data = churnm)
```

```python
sns.lmplot(x = 'lasttrans', y = 'age', data = churnm, hue = "gender")
```

```python
sns.lmplot(x = 'lasttrans', y = 'age', data = churnm, col = "gender")
```

```python
sns.heatmap(churnm.isnull(), yticklabels = False)
```
